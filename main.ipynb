```python
#Добавление библиотек
%matplotlib inline
import matplotlib.pyplot as plt
from openpyxl import load_workbook
```

```python
#Добавление рабочего листа и создание переменных
wb = load_workbook('students.xlsx')
wb_active = wb.active
msrednee = 0
fsrednee = 0
isrednee = 0
```

```python
#Высчитывание средних значений
for i in range(2,109):
    value = int(wb_active.cell(row = i, column = 2).value)
    msrednee += value
print(msrednee / 108)
for i in range(2,109):
    value = int(wb_active.cell(row = i, column = 3).value)
    fsrednee += value
print(fsrednee / 108)
for i in range(2,109):
    value = int(wb_active.cell(row = i, column = 4).value)
    isrednee += value
print(isrednee / 108)
```

```python
#Создание списков с предметами и средними оценками
math = msrednee/108
phys = fsrednee/108
inf =  isrednee/108
objects = ["Математика", "Физика", "Информатика"]
marks = [math, phys, inf]

plt.bar(objects, marks)
plt.title("Средние баллы по предметам: ")
plt.xlabel("Предметы")
plt.ylabel("Оценки")
plt.grid(alpha=0.3, axis='y')
plt.ylim(0,10)
```

```python
#Добавление результатов в students_result
results = {'A':math,'B':phys,'C':inf}
wb = load_workbook('students_result.xlsx')
wbact = wb.active
wbact.cell(row=2, column=1, value=math)
wbact.cell(row=2, column=2, value=phys)
wbact.cell(row=2, column=3, value=inf)

wb.save('students_result.xlsx')
```
